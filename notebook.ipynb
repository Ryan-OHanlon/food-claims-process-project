## Data Discovery and Visualization

### How does the number of claims differ across locations?

There are four possible locations included in this data. the Sao Luis location has the most amount of claims, with Recife being the second. The biggest difference of the amount of claims between the Sao Luis location and the Natal location is 8 claims. This shows that each office is handling about the same amount of claims.

![Count by location](Count_ClaimsLocation.png)

### What is the distribution of time to close claims?

While each location is handling about the same number of claims, we need to look at the distribution of time it takes to close claims. Looking at each claim, we can see that most claims are closed within 1000 days, while the remaining claims range between 1000 and 2000 days to close. There are a couple outliers where claims have taken about and over 3000 days to close the claim. It is important to find out if the outliers where the time to close claims are across all locations or held by a single location.

![Range of claims by Time to close](Range_ClaimsTimetoclose.png)

### How does the average time to close claims differ by location?

Finally, we want to combine the location field with the time to close field and see how long each location takes to close claims.

When looking at each location individually, we can see that the Natal location is able to close claims the fastest compared to the other locations when looking at the average, as well as the interquartile range. Though it should be noted that Natal has the least amount of claims compared to the other locations.

In contrast, the Sao Luis location has the most amount of claims to close but also takes the longest and shortest amount of time to close claims based on the whiskers showing the maximum and minimum time to close claims. To reduce the amount of time it takes to close claims the legal team should investigate the Sao Luis location to determine why the time to close claims has the largest variance. 

The other locations each have their own outliers each having claims that have taken over 2000 days to close to see what impacted closing the claim by the cause or the number of individuals linked to that claim.
